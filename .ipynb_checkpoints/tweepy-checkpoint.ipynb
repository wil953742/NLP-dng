```python
r = re.compile(' ')
df_root = df.loc[[bool(r.match(word)) for word in df['Word']], ['Word', 'Topic']]
df_word = df.loc[[not bool(r.match(word)) for word in df['Word']], ['Word', 'Topic']]

df_root.insert(loc=2, column="Count", value=0)
df_word.insert(loc=2, column="Count", value=0)

df_root = df_root.set_index('Word')
df_word = df_word.set_index('Word')

# 앞에 띄어쓰기 제거해야 함
df_root.head()
#df_word.head()
```

```python
from nltk.tokenize import word_tokenize, RegexpTokenizer

vocab = {}
#temp = ms_user[8].text
#for twt in ms_user:
#    word_tokens = word_tokens + word_tokenize(twt.text)
#tokenizer1 = RegexpTokenizer("\s+|#|@|$|%|^|&|\*|\(|\)|\!|\?|<|>|:|;|~|\.|,", gaps=True)
tokenizer = RegexpTokenizer("[\w]+")

for twt in ms_user:
    words = [ word.lower() for word in tokenizer.tokenize(twt.text) if len(word) >= 2 ]
    for word in words:
        if word not in vocab:
            vocab[word] = 0
        vocab[word] += 1
        
for key, value in vocab.items():
    print(key, ' : ', value)
```

```python
for key in vocab.keys():
    if key in df_word.index:
        df_word.loc[key, 'Count'] += 1
        
df_word_cnt = df_word[df_word['Count'] > 0]
df_word_cnt
```
